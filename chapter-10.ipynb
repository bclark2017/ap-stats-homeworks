### b

```r
fit <- lm(spending ~ year, data=nsf)
plot(spending ~ year, data=nsf)
abline(fit, col="red")
```

### c

```r
nsf
y.hat <- makeFun(fit)
r1 <- nsf$spending[1] - y.hat(year=1999)
r2 <- nsf$spending[2] - y.hat(year=2000)
r3 <- nsf$spending[3] - y.hat(year=2001)

r1
r2
r3

sqrt((r1^2 + r2^2 + r3^2) / 1)
```

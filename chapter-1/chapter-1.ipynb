```r
damage <- read.csv("./ta01_005.csv", header=TRUE)
```

```r
# a.
# five number summary plus the mean
summary(damage)
```

Here you can see that the difference between the minimum and the first quartile is only 2.2. The difference betwen the maximum and the 3rd quartile, however, is almost 50. This points to a strong right skew.

```r
# b.
is.outlier <- function(data.set) {
    q1 <- quantile(data.set, 0.25)
    q3 <- quantile(data.set, 0.75)
    iqr <- q3 - q1
    return (data.set > 1.5 * iqr + q3 || data.set < q1 - 1.5 * iqr)
}

damage$outlier <- is.outlier(damage$damage)
damage$outlier
```

The vector now attached to the data frame contains all values of false, indicative that none of the data are considered outliers according to the 1.5 $\cdot$ IQR rule

```r
print(paste("mean:", mean(damage$damage)))
print(paste("median:", median(damage$damage)))
```

Because the data are so right skewed, the mean is much greater than the median. This is because mean is much less resistive of a statistic.

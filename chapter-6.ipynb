#### A helpful confidence interval function

```r
require(mosaic)
meanconfint = function (x, sigma, level = 0.95, ...) {
    s = sigma / sqrt(length(x))
    mu = mean(x)
    z = qnorm(1 - (1 - level)/2)
    out = c(mu, mu - z * s, mu + z * s)
    names(out) = c("mean", "lower", "upper")
    return(out)
}
```

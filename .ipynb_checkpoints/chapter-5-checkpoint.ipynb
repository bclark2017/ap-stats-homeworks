### 57.

Because $X$ is approx. Normal, $\bar{x}$ is also approx. Normal.

$\mu = 0.2\%; \sigma = 20.6\%$

If $n = 45$, what are $P(\bar{x} > 15 \%)$ and $P(\bar{x} < 5 \%)$?

Well, $\bar{x}$ has $N(\mu, \frac{\sigma}{\sqrt{n}})$, so

```r
mu <- 9.2
sigma <- 20.6
n <- 45
sigma.xbar <- sigma / (sqrt(n))

1 - pnorm(15, mu, sigma.xbar)
```

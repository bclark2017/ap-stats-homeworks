```r
expected <- function(x) {
    return (as.array(margin.table(x, 1)) %*% t(as.array(margin.table(x, 2))) / margin.table(x))
}

df <- function(x) {
    if (nrow(x) == 1) {
        return (ncol(x) - 1)
    } else {
        return ((nrow(x) - 1) * (ncol(x) - 1))
    }
}
```
